### 2.1 Работа с пропусками

#### 2.1.1 Поиск и анализ пропущенных в датафрейме значений

Найдем и изучим пропущенные значения в столбцах датафрейма. Определим, в каких столбцах есть пропуски.

```python
data.isna().sum() # определим количество пропусков
```

```python
(
    round
    (
        data.isna().mean()*100
    )
    .to_frame()
    .style.background_gradient('coolwarm')
) # определим количество пропусков в %
```

Оценим количество пропусков в датафрейме, построив горизонтальный столбчатый график.

```python
(
    data.isna() # определим пропущенные значения в датафрейме
     .agg(['sum', lambda x: round((x.sum() / x.count())*100, 1)]) # применим несколько функций к столбцам датафрейма
     .transpose() # транспонируем датафрейм
     .rename(columns={'sum': 'кол-во_пропусков', '<lambda>': 'процент_пропусков'}) # изменим названия столбцов
     .sort_values(by='процент_пропусков') # выполним сортировку
     .plot(kind='barh', y='процент_пропусков', figsize=(19, 6), fontsize = 14, grid=True) #построим горизонтальный столбчатый график
);
```

Больше всего пропусков в данных с указанием является ли недвижимость в объявлении аппартаментами, больше 60% пропусков в данных о близости к парку или водоему, почти в половине объявлений не указано наличие балконов, а в 40% - высоты потолков, также в каждом пятом случае не удалось автоматически поддянуть картографических данные с расстояниями до центра, аэропорта, количества парков и водоемов в радиусе 3 км, значительно меньше пропусков в данных с указанием площади кухни и жилой площади, а также с информацией о количестве дней от публикации до снятия объявления, название населённого пункта пропущенно в 0,2% случаев.

#### 2.1.2 Замена пропусков в столбце с площадью кухни

Общее количество пропусков в столбце с площадью кухни:

```python
data['kitchen_area'].isna().sum()
```

Выведем гистограмму и описательную статистику:

```python
data['kitchen_area'].hist(bins=100, figsize=(16, 3)); # гистограмма
data['kitchen_area'].describe() # описательная статистика
```

Во-первых, студии отличаются отсутствием кухни, поэтому найдем соответсвующие пропуски и заменим их на 0:

```python
data.loc[data['studio'] == True, 'kitchen_area'] = 0
```

Далее заменим пропуски в соответсвии с общей площадью квартиры. Для этого сначала проведем категоризацию квартир по площади, на гистограмме определим сколько категорий лучше ввести. 

```python
data['total_area'].hist(bins=50, figsize=(16, 3), range=(0, 400));
```

Добавим разбиение общей площади квартир на 5 категорий:
* менее 40 кв.м;
* от 40 до 80 кв.м;
* от 80 до 120 кв.м;
* от 120 до 200 кв.м;
* более 200 кв.м

```python
data['total_area_cat'] = (pd.cut
                          (data['total_area'], 
                           bins=[0, 40, 80, 120, 200, 1000], 
                           labels=['< 40', '40-80', '80-120', '120-200', '> 200'], 
                           right=False)
) # категоризация данных по общей площади квартиры
```

Заменим пропуски средним значением по площади кухонь в соответсвии с проведенной категоризацией по общей площади квартиры:

```python
for n in data['total_area_cat'].unique():
    data.loc[(data['total_area_cat'] == n) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[(data['total_area_cat'] == n), 'kitchen_area'].mean()
```

Проверим отсутствие пропусков:

```python
data['kitchen_area'].isna().sum()
```

Проверим изменения в гистограмме и описательной статистике:

```python
data['kitchen_area'].hist(bins=100, figsize=(16, 3));
data['kitchen_area'].describe()
```

Существенных искажений в данные не внесено, переходим к следующему столбцу.

#### 2.1.5 Замена пропусков в столбце с информацией о жилой площади:

Общее количество пропусков в столбце с информацией о жилой площади:

```python
data['living_area'].isna().sum()
```

```python
data['living_area'].hist(bins=100, figsize=(16, 3));
data['living_area'].describe()
```

На гистограмме видны пики для  студий и маленьких по площади однушек, а также двух- и трехкомнатных квартир.

При одинаковой общей площади жилая площадь студий будет примерно одинакова, поэтому эти данные заполнить средним значением.

```python
for n in data['total_area_cat'].unique():
    data.loc[(data['total_area_cat'] == n) & (data['studio'] == True) & (data['living_area'].isna()), 'living_area'] = \
    data.loc[(data['total_area_cat'] == n) & (data['studio'] == True), 'living_area'].mean()
```

Так как дома строятся в основном по утвержденным типовым проектам, то далее заполним пропущенные значения жилой площади квартир на основании количества комнат и общей площади: 

```python
for n in data['rooms'].unique():
    for m in data['total_area_cat'].unique():
        data.loc[(data['rooms'] == n) & (data['total_area_cat'] == m) & (data['living_area'].isna()), 'living_area'] = \
        data.loc[(data['rooms'] == n) & (data['total_area_cat'] == m), 'living_area'].mean()
```

#### 2.1.12 Выводы

Оценим количество оставшихся пропусков в датафрейме:

```python
(data.isna() # определим пропущенные значения в датафрейме
 .agg(['sum', lambda x: f'{x.sum() / x.count():.1%}']) # применим несколько функций к столбцам датафрейма
 .transpose() # транспонируем датафрейм
 .rename(columns={'sum': 'кол-во_пропусков', '<lambda>': 'процент_пропусков'}) # изменим названия столбцов
 .sort_values(by='процент_пропусков') # выполним сортировку
)
```

```python
# создаем функцию, которая будет исправлять неправильные значения
def fix_ceiling_height(value):
    try:
        if value < 100 and value >= 20: # если значение больше 20, то вероятно забыли поставить точку/запятую
            return value / 10  # делим на 10, чтобы получить метры
        elif value >= 100:  
            return value / 100  # делим на 100, чтобы получить метры
        else:
            return value  # если значение правильное, возвращаем его без изменений
    except:
        pass
# применяем функцию к столбцу ceiling_height
data['ceiling_height'] = data['ceiling_height'].apply(fix_ceiling_height)
```

### 4.4  Определение средней цены квадратного метра

```python
data_top = data.pivot_table(index='locality_name', values=['last_price', 'total_area'], aggfunc=['count', lambda x: round(x.median())])
data_top['m'] = round(data_top['<lambda>', 'last_price'] / data_top['<lambda>', 'total_area'])
data_top.drop(('count', 'total_area'), axis=1, inplace=True)
data_top.columns = ['количество объявлений', 'средняя цена', 'средняя общая площадь', 'стоимость кв.м']
data_top = data_top.rename_axis('населеннный пункт')
data_top = data_top.sort_values(by='количество объявлений', ascending=False)[0:10]
data_top
```
